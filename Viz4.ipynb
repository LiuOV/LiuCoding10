```python
import library
import pandas as pd
import geopandas as gpd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns


import numpy as np
%matplotlib inline
```

```python
PercentData = library.loadList("global-plastic-fate.csv")
LineData = library.loadList("global-plastics-production.csv")
```

```python
discarded = []
recycled = []
incinerated = []

yearlydata = []
linedata=[]

for i in range (1,len(PercentData)): 
    if(PercentData[i][0] == "Discarded"):
        discarded += [float(PercentData[i][3])]
    if(PercentData[i][0] == "Recycled"):
        recycled += [float(PercentData[i][3])]
    if(PercentData[i][0] == "Incinerated"):
        incinerated += [float(PercentData[i][3])]
        
for i in range(1,len(LineData)):
    linedata += [float(LineData[i][3])]
        
for i in range(0,len(discarded)):
    print(discarded[i])
for i in range(0,len(recycled)):
    print(recycled[i])
for i in range(0,len(incinerated)):
    print(incinerated[i])
for i in range(0,len(linedata)):
    print(linedata[i])
```

```python
df2 = pd.DataFrame({"Waste":linedata},index=range(1950,2016))
df = pd.DataFrame({  "Discarded":discarded, "Recycled":recycled, "Incinerated":incinerated }, index=range(1980,2016))
df
```

```python
#data_perc = df.divide(df.sum(axis=1), axis=0)
sns.set_theme()



fig, ax = plt.subplots(2,1,figsize=(10,16))

ax[1].stackplot(range(1980,2016), df.Discarded, df.Recycled, df.Incinerated, labels=['Discarded','Recycled','Incinerated'],colors=["#9bd0d1","#62aeb2","#46a4ac"],alpha=0.8)
ax[0].fill_between(range(1950,2016),df2.Waste,alpha=0.8)
ax[0].set_title("Yearly Plastic Production Globally (1950-2015)",pad=20)
ax[0].set_xlabel("Year",labelpad = 10)
ax[0].set_ylabel("Waste (Millons of Tonnes)",labelpad = 10)

ax[0].tick_params(labelright=True)

plt.legend(loc='lower left')

plt.title('Percentage of Plastic Waste Fate (1980-2015)',pad=20)
ax[1].set_xlabel("Year",labelpad = 10)
ax[1].set_ylabel("Percentage (%)",labelpad = 10)
plt.tick_params(labelright=True)
plt.show()
```
