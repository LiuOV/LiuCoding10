```python
import library
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import numpy as np

def sortKey(b):
    return b[2]

count = -2

Plasticdata = library.loadList("plastic-waste-generation-total.csv")
gdpData = library.loadList("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_3159040.csv")

wastepercapita = []

col = {"countrycode":0,"waste":3,"population":1}

for x in range (0,len(Plasticdata)):                        #Check for country code in first dataset
    temp = (str(Plasticdata[x][col["countrycode"]]))        #Store country name for reference

    for i in range(0,len(gdpData)):                             
        if(temp == (str((gdpData[i][col["countrycode"]])))):   
            temp2 = (float(Plasticdata[x][col["waste"]]))    
            temp3 = (str(gdpData[i][-1]))
            
            while(len(temp3) < 4 and temp3 != "NY.GDP.PCAP.CD"):
                temp3 = (str(gdpData[i][count]))
                print(str((gdpData[i][col["countrycode"]])) + " " + temp3)
                count = count - 1
            
            
            count = -1
            if(temp3 != "NY.GDP.PCAP.CD"):
                temp3 = float(temp3)
                
    
            
                wastepercapita += [[str(Plasticdata[x][col["countrycode"]]),float(temp2/temp3),temp3]]    

            

wastepercapita.sort(key=sortKey, reverse = True)

for n in range(0,len(wastepercapita)):  
    print(wastepercapita[n])
```

```python
df = pd.DataFrame(wastepercapita, columns=['Country', 'Waste', 'Population'])
df
```
