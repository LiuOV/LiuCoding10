```python
import libaray
import pandas as pd
import geopandas as gpd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.patches as mpatches

import numpy as np
%matplotlib inline
```

```python
def sortKey(b):
    return b[2]

LatLong = libaray.loadList("PlasticMarinePollutionGlobalDataset.csv")
```

```python
ProcessedLatLong = []
ProcessedLatLongCD2 = []
ProcessedLatLongCD3 = []
ProcessedLatLongCD4 = []

colours = []
coloursCD2 = []
coloursCD3 = []
coloursCD4 = []


for x in range (2,len(LatLong)): 
    temp = str(LatLong[x][4])
    #print (temp)
    if(temp != "" or float(temp == 0.00) ):
        ProcessedLatLong += [[str(LatLong[x][1]),str(LatLong[x][2]),float(LatLong[x][4])]]
            
    else:
        ProcessedLatLong += [[str(LatLong[x][1]),str(LatLong[x][2]),0.00]]
    
```

```python
for n in range(0,len(ProcessedLatLong)):      
    if((ProcessedLatLong[n][2]) < 10):
        colours += ["#00000000"]
    if((ProcessedLatLong[n][2]) >= 10 and (ProcessedLatLong[n][2]) < 100):
        colours += ["#5ea9ff"]
    if((ProcessedLatLong[n][2]) >= 100 and (ProcessedLatLong[n][2]) < 1000):
        colours += ["#5effb7"]
    if((ProcessedLatLong[n][2]) >= 1000 and (ProcessedLatLong[n][2]) < 10000):
        colours += ["#ffe45e"]
    if((ProcessedLatLong[n][2]) >= 10000 and (ProcessedLatLong[n][2]) < 500000):
        colours += ["#ff5e5e"]
```

```python
#count density 2        
for x in range (2,len(LatLong)): 
    temp2 = str(LatLong[x][5])
    #print (temp2)
    if(temp2 != "" or float(temp2 == 0.00) ):
        ProcessedLatLongCD2 += [[str(LatLong[x][1]),str(LatLong[x][2]),float(LatLong[x][5])]]
            
    else:
        ProcessedLatLongCD2 += [[str(LatLong[x][1]),str(LatLong[x][2]),0.00]]
    
for n in range(0,len(ProcessedLatLongCD2)):      
    if((ProcessedLatLongCD2[n][2]) < 10):
        coloursCD2 += ["#00000000"]
    if((ProcessedLatLongCD2[n][2]) >= 10 and (ProcessedLatLongCD2[n][2]) < 100):
        coloursCD2 += ["#5ea9ff"]
    if((ProcessedLatLongCD2[n][2]) >= 100 and (ProcessedLatLongCD2[n][2]) < 1000):
        coloursCD2 += ["#5effb7"]
    if((ProcessedLatLongCD2[n][2]) >= 1000 and (ProcessedLatLongCD2[n][2]) < 10000):
        coloursCD2 += ["#ffe45e"]
    if((ProcessedLatLongCD2[n][2]) >= 10000 and (ProcessedLatLongCD2[n][2]) < 500000):
        coloursCD2 += ["#ff5e5e"]    
```

```python
#count density 3       
for x in range (2,len(LatLong)): 
    temp3 = str(LatLong[x][6])
    #print (temp3)
    if(temp3 != "" or float(temp3 == 0.00) ):
        ProcessedLatLongCD3 += [[str(LatLong[x][1]),str(LatLong[x][2]),float(LatLong[x][6])]]
            
    else:
        ProcessedLatLongCD3 += [[str(LatLong[x][1]),str(LatLong[x][2]),0.00]]
    
for n in range(0,len(ProcessedLatLongCD3)):      
    if((ProcessedLatLongCD3[n][2]) < 10):
        coloursCD3 += ["#00000000"]
    if((ProcessedLatLongCD3[n][2]) >= 10 and (ProcessedLatLongCD3[n][2]) < 100):
        coloursCD3 += ["#5ea9ff"]
    if((ProcessedLatLongCD3[n][2]) >= 100 and (ProcessedLatLongCD3[n][2]) < 1000):
        coloursCD3 += ["#5effb7"]
    if((ProcessedLatLongCD3[n][2]) >= 1000 and (ProcessedLatLongCD3[n][2]) < 10000):
        coloursCD3 += ["#ffe45e"]
    if((ProcessedLatLongCD3[n][2]) >= 10000 and (ProcessedLatLongCD3[n][2]) < 500000):
        coloursCD3 += ["#ff5e5e"]
```

```python
#count density 4        
for x in range (2,len(LatLong)): 
    temp4 = str(LatLong[x][7])
    #print (temp4)
    if(temp4 != "" or float(temp4 == 0.00) ):
        ProcessedLatLongCD4 += [[str(LatLong[x][1]),str(LatLong[x][2]),float(LatLong[x][7])]]
            
    else:
        ProcessedLatLongCD4 += [[str(LatLong[x][1]),str(LatLong[x][2]),0.00]]
    
for n in range(0,len(ProcessedLatLongCD4)):      
    if((ProcessedLatLongCD4[n][2]) < 10):
        coloursCD4 += ["#00000000"]
    if((ProcessedLatLongCD4[n][2]) >= 10 and (ProcessedLatLongCD4[n][2]) < 100):
        coloursCD4 += ["#5ea9ff"]
    if((ProcessedLatLongCD4[n][2]) >= 100 and (ProcessedLatLongCD4[n][2]) < 1000):
        coloursCD4 += ["#5effb7"]
    if((ProcessedLatLongCD4[n][2]) >= 1000 and (ProcessedLatLongCD4[n][2]) < 10000):
        coloursCD4 += ["#ffe45e"]
    if((ProcessedLatLongCD4[n][2]) >= 10000 and (ProcessedLatLongCD4[n][2]) < 500000):
        coloursCD4 += ["#ff5e5e"]
```
